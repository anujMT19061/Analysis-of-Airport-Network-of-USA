```python
import pandas as pd
import numpy as np
import csv  
```

```python
df = pd.read_csv('data/Airports.csv',skipinitialspace=True, usecols=['Origin_airport', 'Destination_airport'])
org = df.Origin_airport
dest = df.Destination_airport

res = pd.concat([org, dest], axis=0)
res = res.unique()
res = res.tolist()
# print(res)


df1 = pd.read_csv('data/Locations.csv',skipinitialspace=True, usecols=['Address'])
add = df1.Address

add = add.tolist()
# print(add)

intersection  = list(set(res) & set(add)) 
print(len(intersection))

# print(intersection)
```

```python
def AirportsModified():
    # field names  
    fields = ['Origin_airport','Destination_airport','Origin_city','Destination_city','Passengers','Flights'] 
    airports = pd.read_csv('data/Airports.csv',skipinitialspace=True, usecols=fields)

    # name of csv file  
    filename = "data/AirportsModified.csv"

    # writing to csv file  
    with open(filename, 'w', newline='') as csvfile:  
        # creating a csv writer object  
        csvwriter = csv.writer(csvfile)  
            # writing the fields  
        csvwriter.writerow(fields)

        for index, row in airports.iterrows(): 
            if(row['Origin_airport'] in intersection and row['Destination_airport'] in intersection):
                csvwriter.writerow(row) 
```

```python
# us_states_covid19_daily
def uniqueSates():
    df = pd.read_csv('data/AirportsModified.csv',skipinitialspace=True, usecols=['Origin_airport','Destination_airport','Origin_city', 'Destination_city','Passengers','Flights'])
    org = df.Origin_city
    dest = df.Destination_city
    
    fields = ['Origin_airport','Destination_airport','Origin_city','OCcode','Destination_city','Dccode','Passengers','Flights']
    #split()
    
    occode = []
    dccode = []
    
    occity = []
    dccity = []
    
    filename = "data/AMStates.csv"
    
    with open(filename, 'w', newline='') as csvfile:  
    # creating a csv writer object  
        csvwriter = csv.writer(csvfile)      
    
        csvwriter.writerow(fields)
    
        for index, row in df.iterrows(): 
            OC = row['Origin_city']
            DC = row['Destination_city']
            x,y = OC.split(', ')
            a,b = DC.split(', ')

            occity.append(x)
            occode.append(y)

            dccity.append(a)
            dccode.append(b)

            newrow = [row['Origin_airport'],row['Destination_airport'],x,y,a,b,row['Passengers'],row['Flights']]

            csvwriter.writerow(newrow)
        
    Unqstates  = occode + dccode
    Unqstates = list(set(Unqstates))
    print(len(Unqstates))
    
    
    return Unqstates
```

```python
def CovidStates():
    df = pd.read_csv('data/us_states_covid19_daily.csv',skipinitialspace=True, usecols=['state','positive'])
    
    fields = ['state','positive']
    
    UqSt = uniqueSates()
    print(UqSt)
    print(len(UqSt))
    
    filename = "data/covidfinal.csv"

    # writing to csv file  
    with open(filename, 'w', newline='') as csvfile:  
        # creating a csv writer object  
        csvwriter = csv.writer(csvfile)  
            # writing the fields  
        csvwriter.writerow(fields)

        for index, row in df.iterrows():
            if(row['state'] in UqSt):
                csvwriter.writerow(row)  
```
